```python
import pandas as pd
from sklearn.datasets import load_boston
bunch = load_boston()
y = bunch.target
print(y)
X = pd.DataFrame(bunch.data, columns=bunch.feature_names)
print(X)
enc = ce.BinaryEncoder(cols=['CHAS', 'RAD']).fit(X,y)
```

```python
numeric_dataset = enc.transform(X)
```
